# 8. Multiple Linear Regression Modeling

```python
# Linear regression model creation and evaluation
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

# Define X (Features) and y (Labels) variables
X = por_dataset_2["studytime"]
y = por_dataset_2["G3"]
```

```python
# Function to create linear regression models
def linear_model(df, feature_variables, target_variable):
  X = np.array(df[feature_variables])
  y = np.array(df[target_variable])

  # Splitting the data into training and testing sets
  X_train, X_test, y_train, y_test = train_test_split(X,
                                                      y,
                                                      test_size=0.2,
                                                      random_state=0)

  # Creating and fitting the linear regression model
  lr = LinearRegression()
  lr.fit(X_train, y_train)

  # Making predictions and creating a dataframe for results
  y_pred = lr.predict(X_test)
  dictionary = {}

  for feature in feature_variables:
    dictionary[feature] = X_test[:, feature_variables.index(feature)]

  lr_results = pd.DataFrame.from_dict(dictionary)
  lr_results["Ground Truth"] = y_test
  lr_results["Prediction"] = y_pred

  # Printing model evaluation metrics
  print("Target:",target_variable)
  print("Features:", feature_variables)
  print("The accuracy of the model is:", lr.score(X_test, y_test))
  print("The MAE of the model is:", mean_absolute_error(y_test, y_pred))
  print("The intercept (alpha) is:", lr.intercept_)
  print("The coeficients (betas) are: ", lr.coef_)

  return lr_results.head()
```

```python
# Usage of linear_model with different sets of features
linear_model(por_dataset_2, ["G1", "G2", "studytime"], "G3")
```

```python
linear_model(por_dataset_2, ["G1", "G2", "studytime", "freetime", "failures", "absences"], "G3")
```
